```python
from itertools import product
import random
import numpy as np
```

```python
nb_dices = 5 
nb_faces = 6
```

```python
def get_state(nb_dices, nb_faces): 
    l = [list(x) for x in product(range(nb_faces), repeat = nb_dices)]
    #print(l)
    S = []
    for li in l: 
        #state = [dice_val.count(str(i)) for i in range(nb_face)]
        S.append([li.count((i)) for i in range(nb_faces)])
    
        #state = {i:dice_val.count(str(i+1)) for i in range(nb_face)}
    states, proba = np.unique(S, axis =0, return_counts=True)
    return list(states), proba/np.sum(proba)
```

```python
def get_actions_from_state(s): #s =[2,1,0,0]
    nb_actions = np.prod(s+1)
    #A = np.zeros(nb_actions, nb_face)
    l = []
    for i in range(nb_faces): 
        l.append(np.arange(s[i]+1))
        #l = [0,1,2,3], [0,1,2], [0,1],[0,1]
    cas_possible = product(*l)
    return [list(x) for x in cas_possible]
```

```python
def get_states_from_action(a):
    reroll = nb_dices - np.sum(a)
    Sa, Pa = get_state(reroll, nb_faces)
    
    lS = Sa+np.array(a)
    
    return lS,Pa
```

```python
def get_index_state(s):
    
    return np.argwhere((s==S).sum(axis =1)==nb_faces)[0][0]
```

```python
def reward(s,nb_faces):
    a = np.arange(1,nb_faces+1)
    r = 0
    if list(s).count(1) == 4:
        r = 25
    else:
        if list(s).count(1) == 5: 
            r = 40
        else:
            for i in s: 
                if i == 3:
                    for j in s[i:]: 
                        if j == 2: 
                            r = 25 #full
                        else: 
                            r = np.sum(s*a) #brelan
                if i == 4:
                    r = np.sum(s*a) #carrÃ©
                if i == 5: 
                    r = 50
    return r
```

```python
def optimal_policy(nb_dices, nb_faces): 
    S, P = get_state(nb_dices,nb_faces)
    R = []
    for s in S: 
        R.append(reward(s, nb_faces))
    #print('R0')
    #print(R)
    
    A1_optimal = []
    R1 = []
    for s in S:
        A = get_actions_from_state(s)
        ra = 0 #reward moyenne de cette action
        Ra = []
        for a in A:
            Sa, Pa = get_states_from_action(a)
            ra = 0
            for i,sa in enumerate(Sa):
                ra += R[get_index_state(sa)]*Pa[i]
            Ra.append(ra)
        index = Ra.index(max(Ra))
        R1.append(max(Ra))
        A1_optimal.append(np.sum(A[index]))
        
        
    A2_optimal = []
    R2 = []    
    for s in S:
        A = get_actions_from_state(s)
        ra = 0 #reward moyenne de cette action
        Ra = []
        for a in A:
            Sa, Pa = get_states_from_action(a)
            ra = 0
            for i,sa in enumerate(Sa):
                ra += R1[get_index_state(sa)]*Pa[i]
            Ra.append(ra)
        index = Ra.index(max(Ra))
        R2.append(max(Ra))
        A2_optimal.append(np.sum(A[index]))
    return R,R1,R2,A1_optimal,A2_optimal
```

```python
nb_dices = 5
nb_faces = 6
R,R1,R2,A1_optimal,A2_optimal = optimal_policy(nb_dices,nb_faces)
```
